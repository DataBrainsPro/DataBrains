```python

import pandas as pd
import numpy as np
tsa = pd.read_csv('data/tsa_passenger_throughput.csv', parse_dates=['Date'])
tsa.head()
```

```python
tsa = tsa.rename(columns=lambda x: x.lower().split()[0])
```

```python
tsa_melted = tsa.melt(id_vars='date', var_name='year', value_name='travelers')
tsa_melted.sample(5,random_state=1)
```

```python
tsa_melted = tsa_melted.assign(date = lambda x: pd.to_datetime(x.year + x.date.dt.strftime('-%m-%d')))
tsa_melted.sample(5,random_state=1)
```

```python
tsa_melted.sort_values('date').tail()
```

```python

tsa_melted.sort_values('date').tail()
```

```python
tsaPivot = tsa_melted.query('date.dt.month == 3 and date.dt.day <=10').assign(day_in_march = lambda x: x.date.dt.day)\
.pivot(index='year', columns='day_in_march',values='travelers')
tsaPivot
```
